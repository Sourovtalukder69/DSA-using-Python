```python
# stack is a simple data Structure : The main function of stack is push and pop
# Stack can created by array or linked list

# I am creating stack by linked list

class Node:
    def __init__(self, value):
        self.data = value
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
        self.n = 0

    def is_empty(self):
        if self.top == None:
            return True
        else:
            return False
        
    def push(self, value):
        new_node = Node(value)
        if self.top == None:
            self.top = new_node
        else:
            new_node.next = self.top
            self.top = new_node
        self.n = self.n + 1

    def pop(self):
        if self.top == None:
            return 'Empty Stack'
        else:
            pop_item = self.top.data
            self.top = self.top.next
            self.n = self.n-1
        return pop_item
            
    
    def traverse(self):
        curr = self.top
        while curr != None:
            print(curr.data, end=' ')
            curr = curr.next

    def size(self):
        return self.n
    
    # String Reversal

    def string_reversal(self, string):
        s1 = Stack()
        for i in string:
            s1.push(i)

        return s1.traverse()
    
    # Undo redo in stack operation

    def undo_redo(self, string, ur):
        s1 = Stack()
        s2 = Stack()
        for i in  string:
            s1.push(i)
        for j in ur:
            if j == 'u':
                s2.push(s1.pop())
            else:
                 s1.push(s2.pop())
    
        return s1.show()
    
    def show(self):
        curr = self.top
        s3 = Stack()
        while curr != None:
            s3.push(curr.data)
            curr = curr.next
        return s3.traverse()
    
    def bracket_check(self, brackets):
        list = []
        s1 = Stack()
        for i in brackets:
            s1.push(i)
            list.append(i)
        return s1.traverse()

                
        
    
# ([{[()]}])
    


s = Stack()
s.bracket_check('([{})')
```

```python
# Queues : Queues is a linear data structue it follow the princiapl of fifo:
# First in first Out 

# Ques implementation with linked list 

class Node:
    def __init__(self, value):
        self.data = value
        self.next = None

class Queues:
    def __init__(self):
        self.front = None
        self.rare = None
        self.n = 0

    def enqueue(self, value):
        new_node = Node(value)
        if self.rare == None:
            self.front = new_node
            self.rare = self.front
        else:
            self.rare.next = new_node
            self.rare = new_node
            self.n = self.n + 1

    def dqueue(self):
        if self.front == None:
            return 'Empty Queues'
        else:
            self.front = self.front.next
            self.n = self.n - 1


    def traverse(self):
        curr = self.front
        while curr != None:
            print(curr.data, end=' ')
            curr = curr.next
            

q = Queues()
q.enqueue(10)
q.enqueue(5)
q.enqueue(3)
q.dqueue()
q.dqueue()
q.traverse()
```

```python
def merge_sorted(arr1, arr2):  # sorted array given
    i = 0
    j = 0
    merged = []
    while i<len(arr1) and j<len(arr2):
        if arr1[i] < arr2[j]:
            merged.append(arr1[i])
            i = i+1
        else:
            merged.append(arr2[j])
            j = j+1
    
    while i < len(arr1):
        merged.append(arr1[i])
        i = i + 1
    while j < len(arr2):
        merged.append(arr2[j])
        j = j + 1
    
    return merged

arr1 = [1, 2, 4]
arr2 = [3,5, 7, 8, 9, 11]
merge_sorted(arr1, arr2)


def merge_sort(arr):
    if len(arr) == 1:
        return arr
    mid = len(arr)//2
    left = arr[:mid]
    right = arr[mid:]
    left = merge_sort(left)
    right = merge_sort(right)

    return merge_sorted(left, right)
```
